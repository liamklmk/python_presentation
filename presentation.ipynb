##Berechnung der durchschnittlichen Renditen

```python
t = 'data/tsla.csv'
df = pd.read_csv(t, decimal = '.')
df = df[['Date', 'Close']]
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
df = df[(df['Date'].dt.year >= 2011) & (df['Date'].dt.year <= 2022)]
df['Return in %'] = (np.log(df['Close']) - np.log(df['Close']).shift(1)) * 100
df_avg = df['Return in %'].groupby(df['Date'].dt.year).mean() * 100
df_avg = pd.DataFrame(df_avg).reset_index()
df_avg.head()
```

##Abbildung der durchschnittlichen Renditen

```python
plt.plot(df_avg['Date'], df_avg['Return in %']);
```
